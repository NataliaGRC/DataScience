### Preprocesamiento del Conjunto de Datos

```python
preprocess_data = np.array(df['Temp']).reshape(-1, 1)
```

#### Método de Normalización 

#### En esta sección se va a realizar la normalización de los datos leídos del archivo csv por medio de método automático (uso de librerías de la herramienta) y de forma normal (programación de todas las operaciones).

```python
#Preprocesamiento Mediante Normalización Automática

norm = MinMaxScaler()
norm_df = norm.fit_transform(preprocess_data)
norm_df = pd.DataFrame(norm_df, columns=['Temp'])#### Metodo de Normalizacion
norm_df["Date"] = df["Date"]

print("Resultado de la Normalización")
print(norm_df)
```

```python
#Preprocesamiento Mediante Normalización de forma manual
def CalcularZ(x, Min, Max): 
    return (x-Min) / (Max - Min)

def ObtenerMayor(lista):     
    DatoMayor =0
    for i in lista: 
        if i > DatoMayor:
            DatoMayor =i        
    return DatoMayor

def ObtenerMenor(lista): 
    DatoMenor =0
    for i in lista: 
        if i < DatoMenor:
            DatoMenor =i        
    return DatoMenor

CantidadElementos =0
NormaManual = []
Mayor = ObtenerMayor (preprocess_data)
Menor = ObtenerMenor (preprocess_data)

for i in preprocess_data: 
    NormaManual.append( CalcularZ(i,Menor,Mayor))
    CantidadElementos += 1
  
print(NormaManual)
```

#### Método de Estandarización

```python
#Preprocesamiento Mediante Estandarización automática

stand = StandardScaler().fit(preprocess_data)
stand_df = stand.transform(preprocess_data) 
stand_df = pd.DataFrame(stand_df, columns=['Temp'])
stand_df["Date"] = df["Date"]

print("Resultado de la Estandarizacion")
print(stand_df)
```

```python
#Preprocesamiento Mediante Estandarización manual

EstManual = []


#Calculo de la norma 
def CalcularMedia(lista, CantElementos):     
  sumatoria = 0
  for item in lista:
    sumatoria += item      
  return sumatoria/CantElementos

CalMedia = CalcularMedia (preprocess_data, CantidadElementos)


#Cálculo de la norma 
def DesviacionEstandar(lista, Media, CantElementos):
  SumaCuadrados=0
  for item in lista:
    SumaCuadrados += (item-Media)**2    
  Precalculo = SumaCuadrados / CantElementos
  return mt.sqrt(Precalculo) #Raíz cuadrada de la sumatoria de los cuadrados entre la cantidad de elementos

#Cálculo de la estandarización
def CalcularZE(x, Media, DesviacionE):  
  return (x-Media)/DesviacionE 

DE = DesviacionEstandar (preprocess_data, CalMedia, CantidadElementos)


for i in preprocess_data: 
    EstManual.append(CalcularZE(i,CalMedia,DE))
    
print("Calculo de la Estandarización Manual ")
print(EstManual)
```
